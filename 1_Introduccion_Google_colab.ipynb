## 1. Celdas en `Google Colab`

`Google Colab` nos proporciona un ambiente interactivo para escribir y editar códigos en `Python` utilizando `Jupyter notebook` (además de otros lenguajes de programación).

Los documentos en `Jupyter notebook` se componen de celdas y estas a su vez son `scripts` de `Python` (u otro lenguaje) las cuales pueden ejecutar código o contener texto explicativo:

###1. Celdas de código


####1. Celda de código
Una celda de código ejecuta el `script` que se encuentra dentro de ella. Para ejecutar la celda nos posicionamos sobre esta y precionamos `Shift - Enter` o se hacemos `click` sobre el botón `play`.

####2. Añadir celda de código
Para añadir una celda de código:
- Nos posicionamos en la parte del documento en donde deseamos añadirla. 
- Después hacemos `click` en `+ Code`.

```
radio = 12
area = 3.1416 * radio ** 2
print('El círculo tiene un área de', area, 'cm2')
```

###2. Celdas de texto


####1. Celda de texto

- Nos posicionamos en la parte del documento en donde deseamos añadirla.

##3. `GPU`s y `TPU`s en `Google Colab`
Algunos algoritmos de `Machine Learning` como lo son las: 


*   Redes Neuronales Convolucionales (`CNNs`),
*   Redes Neuronales Recursivas (`RNNs`),
*   Redes Generativas Antagónicas (`GANs`),
*   Algoritmos de Aprendizaje Reforzado (`Reinforcement learning`),
*   Entre otros.


representan el estado del arte para una gran cantidad de cantidad de aplicaciones. Sin embargo, para entrenar eficientemente estos modelos, en grandes cantidades de información, se requiere de hardware especializado como lo son las `GPU`s y `TPU`s. 

Una `GPU` (Graphics Processing Unit) es una unidad de procesamiento de gráficos la cual ejecuta operaciones en paralelo. Y una `TPU` es una unidad de procesamiento de `Tensores`.

`GPUs` y `TPUs` son utilizados como aceleradores en algoritmos 
paralelizables. Los podemos ver como `chips` muy especializados que realizan operaciones de manera muy exacta y muy rápida.

El entrenamiento de este tipo de algoritmos en `CPU`s puede tomar días, semanas o incluso meses. Utilizando `GPUs` o `TPUs` lo podemos reducir a minutos u horas.

La buena noticia es que podemos usar este tipo de algoritmos utilizando las `GPU`s y `TPU`s de `Google Colab` de manera gratuita.

Para utilizar este `hardware` en nuestro `notebook` necesitamos habilitarlo:

- Navegamos hasta `Runtime` → `Change runtime type`.
- Seleccionamos `GPU` o `TPU`.

por default está en `None`, corriendo sobre la `CPU` de la máquina virtual.
